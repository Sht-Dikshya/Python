```python
import pandas as pd
from IPython.display import display

# Load data for the 5 companies
sbl = pd.read_csv('data/SBL_2000-01-01_2021-12-31.csv')
nabil = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
shl = pd.read_csv('data/SHL_2000-01-01_2021-12-31.csv')

# Convert 'Date' column to datetime format
for df in [sbl, nabil, ntc, sanima, shl]:
    df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the last 10 years (2012-2021)
start_year = 2012
filtered_data = { 
    "SBL": sbl[sbl["Date"].dt.year >= start_year], 
    "NABIL": nabil[nabil["Date"].dt.year >= start_year], 
    "NTC": ntc[ntc["Date"].dt.year >= start_year], 
    "Sanima": sanima[sanima["Date"].dt.year >= start_year], 
    "SHL": shl[shl["Date"].dt.year >= start_year]
}

# Aggregate data yearly (Mean Close Price & Total Traded Shares)
agg_data = {}
for company, df in filtered_data.items():
    yearly_data = df.groupby(df['Date'].dt.year).agg({'Close Price': 'mean', 'Total Traded Shares': 'sum'}).reset_index()
    yearly_data['Company'] = company  # Add company name column
    agg_data[company] = yearly_data

# Display yearly aggregated data
for company, data in agg_data.items():
    display(f"{company}'s Yearly Aggregated Data:", data)
```

```python
import pandas as pd
from IPython.display import display

# Load data for the 5 companies
sbl = pd.read_csv('data/SBL_2000-01-01_2021-12-31.csv')
nabil = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
shl = pd.read_csv('data/SHL_2000-01-01_2021-12-31.csv')

# Convert 'Date' to datetime format
for df in [sbl, nabil, ntc, sanima, shl]:
    df['Date'] = pd.to_datetime(df['Date'])

# Function for data cleaning
def clean_data(df, company_name):
    # Check for missing values
    missing_values = df[df.isnull().any(axis=1)]
    display(f"Missing values in {company_name} before cleaning:", missing_values.head())

    # Set 'Date' as index for easier handling
    df.set_index('Date', inplace=True)

    # Forward-fill missing values
    df.fillna(method='ffill', inplace=True)

    # Reset index
    df.reset_index(inplace=True)

    # Display cleaned data
    display(f"Cleaned {company_name} data:", df.head())

# Apply cleaning function to all datasets
clean_data(sbl, "SBL")
clean_data(nabil, "NABIL")
clean_data(ntc, "NTC")
clean_data(sanima, "Sanima")
clean_data(shl, "SHL")
```

# Trend Analysis
**Explanation:**
1. Data Loading and Preprocessing:
- Data is loaded from the CSV files into DataFrames.
- Dates are converted to datetime format for easy plotting.

2. Moving Average Calculation:
- We calculate both the 50-day and 200-day Simple Moving Averages (SMA) for each company's closing price.

3. Trend Detection and Plotting:
- Bullish Trend: When the price is above the moving average, indicating upward momentum.
- Bearish Trend: When the price is below the moving average, indicating downward momentum.


- The function plot_trend takes care of plotting:\
    Close Price as a solid purple line.\
    50-day SMA and 200-day SMA as blue and pink lines, respectively, with varying transparency (alpha).\
    Bullish trend (50-day SMA above the 200-day SMA) is indicated with dots for the 50-day SMA and dashes for the 200-day SMA.\
    Bearish trend (50-day SMA below the 200-day SMA) is indicated with dashes for the 50-day SMA and dots for the 200-day SMA.

4. Plot Settings:

    Titles, labels, and legends are added for clarity.
    The grid is enabled for better visibility.

 This code will generate individual plots for each company showing the trends using different line styles and colors to indicate bullish and bearish patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for each company
sbl_data = pd.read_csv('data/SBL_2000-01-01_2021-12-31.csv')
nabil_data = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc_data = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima_data = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
shl_data = pd.read_csv('data/SHL_2000-01-01_2021-12-31.csv')

# Convert the 'Date' column to datetime
sbl_data['Date'] = pd.to_datetime(sbl_data['Date'])
nabil_data['Date'] = pd.to_datetime(nabil_data['Date'])
ntc_data['Date'] = pd.to_datetime(ntc_data['Date'])
sanima_data['Date'] = pd.to_datetime(sanima_data['Date'])
shl_data['Date'] = pd.to_datetime(shl_data['Date'])

# Calculate 50-day and 200-day moving averages
sbl_data['50_SMA'] = sbl_data['Close Price'].rolling(window=50).mean()
sbl_data['200_SMA'] = sbl_data['Close Price'].rolling(window=200).mean()

nabil_data['50_SMA'] = nabil_data['Close Price'].rolling(window=50).mean()
nabil_data['200_SMA'] = nabil_data['Close Price'].rolling(window=200).mean()

ntc_data['50_SMA'] = ntc_data['Close Price'].rolling(window=50).mean()
ntc_data['200_SMA'] = ntc_data['Close Price'].rolling(window=200).mean()

sanima_data['50_SMA'] = sanima_data['Close Price'].rolling(window=50).mean()
sanima_data['200_SMA'] = sanima_data['Close Price'].rolling(window=200).mean()

shl_data['50_SMA'] = shl_data['Close Price'].rolling(window=50).mean()
shl_data['200_SMA'] = shl_data['Close Price'].rolling(window=200).mean()

# Create a function to plot the trends with different line styles
def plot_trend(data, company_name):
    plt.figure(figsize=(10, 6))

    # Plot Close Price (solid line)
    plt.plot(data['Date'], data['Close Price'], label=f'{company_name} Close Price', color='purple', linestyle='-', linewidth=2)

    # Plot 50-day and 200-day moving averages
    plt.plot(data['Date'], data['50_SMA'], label=f'{company_name} 50-day SMA', color='blue', linestyle='-', alpha=0.6)
    plt.plot(data['Date'], data['200_SMA'], label=f'{company_name} 200-day SMA', color='pink', linestyle='-', alpha=0.6)

    # Check for Bullish or Bearish trends
    if data['50_SMA'].iloc[-1] > data['200_SMA'].iloc[-1]:  # Bullish trend (50-day > 200-day)
        plt.plot(data['Date'], data['50_SMA'], color='blue', linestyle=':', alpha=0.9)
        plt.plot(data['Date'], data['200_SMA'], color='pink', linestyle='--', alpha=0.9)
    else:  # Bearish trend (50-day < 200-day)
        plt.plot(data['Date'], data['50_SMA'], color='pink', linestyle=':', alpha=0.9)
        plt.plot(data['Date'], data['200_SMA'], color='blue', linestyle='--', alpha=0.9)

    # Add titles and labels
    plt.title(f'{company_name} Price and Moving Averages', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price', fontsize=12)
    plt.legend()
    plt.grid(True, alpha= 0.3)
    plt.show()

# Plot for each company
plot_trend(sbl_data, 'SBL')
plot_trend(nabil_data, 'Nabil')
plot_trend(ntc_data, 'NTC')
plot_trend(sanima_data, 'Sanima')
plot_trend(shl_data, 'SHL')
```

# Volatility Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data for each company
sbl_data = pd.read_csv('data/SBL_2000-01-01_2021-12-31.csv')
nabil_data = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc_data = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima_data = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
shl_data = pd.read_csv('data/SHL_2000-01-01_2021-12-31.csv')

# Step 2: Calculate Standard Deviation (Volatility) for each company
def calculate_volatility(data):
    return data['Close Price'].std()

sbl_volatility = calculate_volatility(sbl_data)
nabil_volatility = calculate_volatility(nabil_data)
ntc_volatility = calculate_volatility(ntc_data)
sanima_volatility = calculate_volatility(sanima_data)
shl_volatility = calculate_volatility(shl_data)

# Step 3: Calculate Bollinger Bands (Middle Band, Upper Band, Lower Band)
def bollinger_bands(data, window=20):
    sma = data['Close Price'].rolling(window=window).mean()  # Simple moving average
    std = data['Close Price'].rolling(window=window).std()  # Standard deviation
    upper_band = sma + (2 * std)
    lower_band = sma - (2 * std)
    return sma, upper_band, lower_band

# Calculate Bollinger Bands for each company
sbl_sma, sbl_upper, sbl_lower = bollinger_bands(sbl_data)
nabil_sma, nabil_upper, nabil_lower = bollinger_bands(nabil_data)
ntc_sma, ntc_upper, ntc_lower = bollinger_bands(ntc_data)
sanima_sma, sanima_upper, sanima_lower = bollinger_bands(sanima_data)
shl_sma, shl_upper, shl_lower = bollinger_bands(shl_data)

# Step 4: Compare Volatility Across Companies
volatility_data = {
    'SBL': sbl_volatility,
    'NABIL': nabil_volatility,
    'NTC': ntc_volatility,
    'SANIMA': sanima_volatility,
    'SHL': shl_volatility
}

# Plot the Volatility Comparison
plt.figure(figsize=(10, 6))
plt.bar(volatility_data.keys(), volatility_data.values(), color='skyblue')
plt.title('Stock Price Volatility Comparison (Standard Deviation)')
plt.xlabel('Company')
plt.ylabel('Volatility (Standard Deviation)')
plt.show()

# Step 5: Plot Bollinger Bands for SBL as an example (you can do similarly for other companies)
plt.figure(figsize=(10, 6))
plt.plot(sbl_data['Date'], sbl_data['Close Price'], label='SBL Close Price', color='blue')
plt.plot(sbl_data['Date'], sbl_upper, label='SBL Upper Bollinger Band', linestyle='--', color='red')
plt.plot(sbl_data['Date'], sbl_lower, label='SBL Lower Bollinger Band', linestyle='--', color='red')
plt.title('SBL Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# You can repeat the same plotting process for Nabil, NTC, Sanima, and SHL if needed.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate daily returns (percentage change)
def calculate_returns(data):
    # Ensure the 'Close Price' column exists
    if 'Close Price' not in data.columns:
        raise ValueError("'Close Price' column is missing from the data.")
    
    data['Return'] = data['Close Price'].pct_change()
    return data

# Add volatility (standard deviation) and Bollinger Bands to the data
def calculate_volatility_and_bollinger_bands(data):
    # Ensure that the returns are calculated before this step
    data = calculate_returns(data)
    
    # Calculate 20-day moving average
    data['20_MA'] = data['Close Price'].rolling(window=20).mean()
    
    # Calculate 20-day standard deviation (volatility)
    data['20_STD'] = data['Close Price'].rolling(window=20).std()
    
    # Calculate Bollinger Bands
    data['Upper_Band'] = data['20_MA'] + (2 * data['20_STD'])
    data['Lower_Band'] = data['20_MA'] - (2 * data['20_STD'])
    
    # Calculate standard deviation (volatility) for the entire dataset
    data['Volatility'] = data['Return'].std()
    return data

# Example: Load your data (replace with actual data)
# sbl_data = pd.read_csv('sbl_data.csv')  # Example, replace with actual data loading method

# Assuming sbl_data is already loaded and has a 'Close Price' column

# Apply volatility and Bollinger Bands calculations
sbl_data = calculate_volatility_and_bollinger_bands(sbl_data)
nabil_data = calculate_volatility_and_bollinger_bands(nabil_data)
ntc_data = calculate_volatility_and_bollinger_bands(ntc_data)
sanima_data = calculate_volatility_and_bollinger_bands(sanima_data)
shl_data = calculate_volatility_and_bollinger_bands(shl_data)

# Plot the Bollinger Bands and Close Price for each company
def plot_volatility(data, company_name):
    plt.figure(figsize=(10, 6))
    
    # Plot Close Price
    plt.plot(data['Date'], data['Close Price'], label=f'{company_name} Close Price', color='blue', linewidth=2)
    
    # Plot Bollinger Bands
    plt.plot(data['Date'], data['Upper_Band'], label=f'{company_name} Upper Band', color='green', linestyle='--', alpha=0.6)
    plt.plot(data['Date'], data['Lower_Band'], label=f'{company_name} Lower Band', color='red', linestyle='--', alpha=0.6)
    
    # Plot 20-day moving average
    plt.plot(data['Date'], data['20_MA'], label=f'{company_name} 20-day MA', color='orange', linestyle='-', alpha=0.8)
    
    # Add titles and labels
    plt.title(f'{company_name} Volatility with Bollinger Bands', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot for each company
plot_volatility(sbl_data, 'SBL')
plot_volatility(nabil_data, 'Nabil')
plot_volatility(ntc_data, 'NTC')
plot_volatility(sanima_data, 'Sanima')
plot_volatility(shl_data, 'SHL')

# Compare volatility across companies (Standard Deviation of Returns)
def compare_volatility(*companies):
    volatility_data = {}
    for company in companies:
        volatility_data[company.name] = company['Volatility'].iloc[-1]  # Volatility based on returns
    return volatility_data

# Compare volatility for the selected companies
volatility_comparison = compare_volatility(sbl_data, nabil_data, ntc_data, sanima_data, shl_data)
print("Volatility Comparison (Standard Deviation of Returns):")
for company, volatility in volatility_comparison.items():
    print(f'{company}: {volatility:.4f}')
```
