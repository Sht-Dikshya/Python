# NumPy Arithmetic Operations
- add()
- subtract()
- multiply()
- divide()
- power()
- mod()

```python
import numpy as np

first_array = np.array([1,3,5,7])
second_array = np.array([2,4,6,8])

# + operator 
result1 = first_array + second_array 
print(" + operator:", result1)
#add() function
result1 = np.add (first_array, second_array )
print(" ADD():", result1)

# - operator 
result1 = first_array - second_array 
print(" - operator:", result1)
#Subtract() function
result1 = np.subtract (first_array, second_array )
print(" SUBTRACT():", result1)

# * operator 
result1 = first_array * second_array 
print(" * operator:", result1)
#multiply() function
result1 = np.multiply (first_array, second_array )
print(" MULTIPLY():", result1)

# / operator 
result1 = first_array / second_array 
print(" / operator:", result1)
#divide() function
result1 = np.divide (first_array, second_array )
print(" Divide():", result1)
```

# Numpy Mathematical and Statistical Functions

```python
numbers= [1,2,3,4]
"""Mathematical Functions"""
sqrt_numbers= np.sqrt(numbers)
print("Square root: ",sqrt_numbers)
square_numbers= np.square(numbers)
print("Square no: ",square_numbers)


"""Statistical Functions"""
students_marks = np.array([38,50,67,10,100,99,98,71,54,341])
mean_marks= np.mean(students_marks)
print("Mean:", mean_marks)

#compute the median of marks

median_marks= np.median(students_marks)
print("Median:", median_marks)

#find the minimum and maximum marks
min_marks= np.min(students_marks)
print("Minimum marks:", min_marks)

max_marks= np.max(students_marks)
print("Maximum marks:", max_marks)
```

# Gaussian Naive Bayes Classifier

- It is a supervised machine-learning algorithm based on Baye's Theorem

- The Naive Bayes classifier assumes that the effect of a particular feature in a class is independent of other features

- For example, a loan applicant is desirable or not depending on his/her income, previous loan and transaction history, age, and location. Even if these features are interdependent, these features are still considered 

    - **P(A | B):** Probability of event A happening given that B is true **(posterior probability)** i.e. probability of event after evidence is seen

    - **P(B | A):** Probability of event B given that A is true **(likelihood)**. i.e. the likelihood that a hypothesis will come true based on the evidence

    - **P(A):** Probability of event A occurring **(prior probability)**. i.e. probability of event before evidence is seen

    - **P(B):** Probability of event B occurring **(marginal probability)** i.e probability of Evidence

**P(A|B)= {P(B|A).P(A)} / P(B)**\
i.e. Posterior= (Likelihood.Prior)/ Marginal

# Deep Learning 

- Lets create a Neural Network with 
    - One input Neuron 
    - One Hidden Layer 
    - One output Neuron 
- Train thios Neural Network to find the pattern in our dataset 
- Out dataset congtains 
    - x = investment on Stock Market 
    - y = Return(=x*3)
- Train The Neural Network and later predict The Return for new investment.

```python
import numpy as np

np.random.seed(42)
x_train= np.random.normal(0,1,(1000,1))
y_train= 3* x_train

# Initialize weights and biases
w1= np.random.randn(1,1)
b1= np.zeros((1, ))
w2= np.random.randn(1,1)
b2= np.zeros((1, ))

def relu(x):
    return np.maximum(0,x)
def relu_derivative(x):
    return (x>0).astype(float)

# Learning rate
lr= 0.01
epochs= 100

# Training loop
for epoch in range(epochs):
    # Forward pass
    hidden_input= np.dot(x_train, w1)+ b1
    hidden_output= relu(hidden_input)   #Hidden layer activation

    output= np.dot(hidden_output,w2)+b2 # Output layer (linear activation)

    # Compute loss(Mean Squared Error)
    loss= np.mean((output-y_train)**2)

    # Backpropagation
    d_loss_output = 2 * (output - y_train)/ len(y_train)     #dL/dy 
    d_w2 = np.dot(hidden_output.T, d_loss_output) # gradient for w2 
    d_b2 = np.sum(d_loss_output, axis = 0)     # gradient for b2

    d_hidden= np.dot(d_loss_output,w2.T)*relu_derivative(hidden_input)
    d_w1= np.dot(x_train.T, d_hidden)   #Gradient for w1
    d_b1= np.sum(d_hidden,axis=0)       # Gradient for b1

    # Update we 
    w1 -= lr * d_w1
    b1 -= lr * d_b1
    w2 -= lr * d_w2 
    b2 -= lr * d_b2

    # Print loss every 100 epochs
    if epoch% 100 == 0:
        print(f"Epoch{epoch},Loss: {loss:.4f}")

# Testing the trained model 
x_test = np.array([[2.0], [7.5], [0.0]])  # Fixing test inputs
hidden_input_test = np.dot(x_test, w1) + b1
hidden_output_test = relu(hidden_input_test)
y_predi = np.dot(hidden_output_test, w2) + b2  # Output layer prediction

# PRINT RESULTS
for i, x_val in enumerate(x_test):
    print(f"Input: {x_val[0]}, Predicted: {y_predi[i][0]:.4f}, Expected: {3*x_val[0]:.4f}")


    
```
