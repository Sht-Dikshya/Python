<b><h4>Introduction</h4></b>
This assignment focuses on analyzing the historical stock prices and trading data from the Nepal Stock Exchange (NEPSE) for five different companies: BFC, NABIL, NTC, Sanima, and SMATA. The goal of this project is to retrieve, clean, process, and visualize stock-related information to gain insights into their performance over a period of time (from 2000 to 2021). The data includes daily closing prices, trading volumes, and stock indices for each company.

<b><h5>Data Collection</h5></b>
For data collection, I retrieved historical stock prices for the five companies from publicly available CSV files. The data includes daily information such as the closing price and the total number of shares traded. The data was collected for a period spanning from January 1, 2000, to December 31, 2021. One of the primary challenges I faced during this phase was handling missing data points due to holidays or days without trading. These missing values can affect the integrity of the analysis and needed to be addressed during the data cleaning process.

<b><h5>Data Cleaning</h5></b>
After gathering the data, the next step was data cleaning. Missing data points were present in several rows due to holidays or non-trading days. To handle this, I applied a method called forward-filling. This technique involves filling missing values by propagating the previous valid data forward. I used this approach for both the "Close Price" and "Total Traded Shares" columns to ensure a consistent data set for analysis.

Additionally, I verified that there were no duplicates or unnecessary columns in the dataset and made sure that the "Date" column was in a proper datetime format. This is critical as accurate date formatting is essential for time-series analysis.

<b><h5>Data Processing</h5></b>
The next step was processing the data. I extracted relevant columns for analysis, specifically the daily closing price and total traded shares for each company. For each company, I also added a new column to indicate the company name for easier identification. This processing was done using the Pandas library in Python. After extracting the daily data, I grouped it by year to calculate the average closing price and the total traded shares per year. This allowed for a high-level comparison between the performance of the companies over time.

<b><h5>Data Visualization</h5></b>
To better understand the trends and comparisons, I used bar charts to visualize the average closing price and total traded shares of each company for the years 2000â€“2021. The bar charts provide a clear visual representation of how each company's stock price and trading volume compares on average across the specified time period.

# Top 5 data(i.e. dates, close price and trade volume) of each company
**BFC, NABIL, NTC, Sanima, and SMATA.**

```python
import pandas as pd
from IPython.display import display

# Load data for the 5 companies
bfc = pd.read_csv('data/BFC_2000-01-01_2021-12-31.csv')
nabil = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
smata = pd.read_csv('data/SMATA_2000-01-01_2021-12-31.csv')

# Extract daily closing prices, trading volumes, and add company name
bfc_processed = bfc[['Date', 'Close Price', 'Total Traded Shares']]
bfc_processed['Company'] = 'BFC'

nabil_processed = nabil[['Date', 'Close Price', 'Total Traded Shares']]
nabil_processed['Company'] = 'NABIL'

ntc_processed = ntc[['Date', 'Close Price', 'Total Traded Shares']]
ntc_processed['Company'] = 'NTC'

sanima_processed = sanima[['Date', 'Close Price', 'Total Traded Shares']]
sanima_processed['Company'] = 'Sanima'

smata_processed = smata[['Date', 'Close Price', 'Total Traded Shares']]
smata_processed['Company'] = 'SMATA'

# Display processed data for each company
display("BFC's data here:", bfc_processed.head())
display("NABIL's data here:", nabil_processed.head())
display("NTC's data here:", ntc_processed.head())
display("Sanima's data here:", sanima_processed.head())
display("SMATA's data here:", smata_processed.head())
```

# Yearly avg data of 5 company (BFC, NABIL, NTC, Sanima, and SMATA.)

```python
import pandas as pd
from IPython.display import display

# Load data for the 5 companies
bfc = pd.read_csv('data/BFC_2000-01-01_2021-12-31.csv')
nabil = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
smata = pd.read_csv('data/SMATA_2000-01-01_2021-12-31.csv')


# Group by year and calculate the average closing price and total traded shares per year
bfc_yearly = bfc_processed.groupby(bfc_processed['Date'].dt.year).agg({'Close Price': 'mean', 'Total Traded Shares': 'sum'}).reset_index()
bfc_yearly['Company'] = 'BFC'

nabil_yearly = nabil_processed.groupby(nabil_processed['Date'].dt.year).agg({'Close Price': 'mean', 'Total Traded Shares': 'sum'}).reset_index()
nabil_yearly['Company'] = 'NABIL'

ntc_yearly = ntc_processed.groupby(ntc_processed['Date'].dt.year).agg({'Close Price': 'mean', 'Total Traded Shares': 'sum'}).reset_index()
ntc_yearly['Company'] = 'NTC'

sanima_yearly = sanima_processed.groupby(sanima_processed['Date'].dt.year).agg({'Close Price': 'mean', 'Total Traded Shares': 'sum'}).reset_index()
sanima_yearly['Company'] = 'Sanima'

smata_yearly = smata_processed.groupby(smata_processed['Date'].dt.year).agg({'Close Price': 'mean', 'Total Traded Shares': 'sum'}).reset_index()
smata_yearly['Company'] = 'SMATA'

# Display the yearly data for each company
display("BFC's yearly data:", bfc_yearly)
display("NABIL's yearly data:", nabil_yearly)
display("NTC's yearly data:", ntc_yearly)
display("Sanima's yearly data:", sanima_yearly)
display("SMATA's yearly data:", smata_yearly)
```

# Checking missing value of SMATA

```python
import pandas as pd
from IPython.display import display

# Load data for the 5 companies
smata = pd.read_csv('data/SMATA_2000-01-01_2021-12-31.csv')

# Convert 'Date' to datetime format
smata['Date'] = pd.to_datetime(smata['Date'])

# Check for missing values in the 'Close Price' and 'Total Traded Shares' columns
missing_value = smata[smata['Close Price'].isnull() | smata['Total Traded Shares'].isnull()]

# Display one missing value example (if available)
display("One missing value example from SMATA data before cleaning:", missing_value.head(1))

# Set 'Date' as the index for easier handling
smata.set_index('Date', inplace=True)

# Forward-fill missing values in SMATA
smata['Close Price'] = smata['Close Price'].fillna(method='ffill')
smata['Total Traded Shares'] = smata['Total Traded Shares'].fillna(method='ffill')

# Reset the index after forward-filling
smata.reset_index(inplace=True)

# Display cleaned SMATA data
display("Cleaned SMATA data:", smata.head())
```

# Merge dataframe of all five company 


```python
import pandas as pd
from IPython.display import display

# Load data for each company
bfc = pd.read_csv('data/BFC_2000-01-01_2021-12-31.csv')
nabil = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
smata = pd.read_csv('data/SMATA_2000-01-01_2021-12-31.csv')

# Convert 'Date' to datetime format for consistency
bfc['Date'] = pd.to_datetime(bfc['Date'])
nabil['Date'] = pd.to_datetime(nabil['Date'])
ntc['Date'] = pd.to_datetime(ntc['Date'])
sanima['Date'] = pd.to_datetime(sanima['Date'])
smata['Date'] = pd.to_datetime(smata['Date'])

# Select relevant columns and rename for clarity
bfc = bfc[['Date', 'Close Price', 'Total Traded Shares', 'Max. Price', 'Min. Price']].rename(
    columns={'Close Price': 'BFC_Close', 'Total Traded Shares': 'BFC_Traded', 'Max. Price': 'BFC_Max', 'Min. Price': 'BFC_Min'})

nabil = nabil[['Date', 'Close Price', 'Total Traded Shares', 'Max. Price', 'Min. Price']].rename(
    columns={'Close Price': 'NABIL_Close', 'Total Traded Shares': 'NABIL_Traded', 'Max. Price': 'NABIL_Max', 'Min. Price': 'NABIL_Min'})

ntc = ntc[['Date', 'Close Price', 'Total Traded Shares', 'Max. Price', 'Min. Price']].rename(
    columns={'Close Price': 'NTC_Close', 'Total Traded Shares': 'NTC_Traded', 'Max. Price': 'NTC_Max', 'Min. Price': 'NTC_Min'})

sanima = sanima[['Date', 'Close Price', 'Total Traded Shares', 'Max. Price', 'Min. Price']].rename(
    columns={'Close Price': 'Sanima_Close', 'Total Traded Shares': 'Sanima_Traded', 'Max. Price': 'Sanima_Max', 'Min. Price': 'Sanima_Min'})

smata = smata[['Date', 'Close Price', 'Total Traded Shares', 'Max. Price', 'Min. Price']].rename(
    columns={'Close Price': 'SMATA_Close', 'Total Traded Shares': 'SMATA_Traded', 'Max. Price': 'SMATA_Max', 'Min. Price': 'SMATA_Min'})

# Merge data using inner join (Keeps only dates available in all 5 datasets)
merged_data = pd.merge(bfc, nabil, on='Date', how='inner')
merged_data = pd.merge(merged_data, ntc, on='Date', how='inner')
merged_data = pd.merge(merged_data, sanima, on='Date', how='inner')
merged_data = pd.merge(merged_data, smata, on='Date', how='inner')

# Sort by Date
merged_data.sort_values('Date', inplace=True)

# Display the first few rows of the merged dataframe
display("Merged Data:", merged_data.head(10))
```

# Bar chart for average close price and Total Traded Shares


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data for the 5 companies
bfc = pd.read_csv('data/BFC_2000-01-01_2021-12-31.csv')
nabil = pd.read_csv('data/NABIL_2000-01-01_2021-12-31.csv')
ntc = pd.read_csv('data/NTC_2000-01-01_2021-12-31.csv')
sanima = pd.read_csv('data/SANIMA_2000-01-01_2021-12-31.csv')
smata = pd.read_csv('data/SMATA_2000-01-01_2021-12-31.csv')

# Convert 'Date' to datetime format
bfc['Date'] = pd.to_datetime(bfc['Date'])
nabil['Date'] = pd.to_datetime(nabil['Date'])
ntc['Date'] = pd.to_datetime(ntc['Date'])
sanima['Date'] = pd.to_datetime(sanima['Date'])
smata['Date'] = pd.to_datetime(smata['Date'])

# Forward-fill missing data for SMATA
smata.set_index('Date', inplace=True)
smata['Close Price'] = smata['Close Price'].fillna(method='ffill')
smata['Total Traded Shares'] = smata['Total Traded Shares'].fillna(method='ffill')
smata.reset_index(inplace=True)

# Compute average Close Price and Total Traded Shares for each company
bfc_avg = bfc['Close Price'].mean()
nabil_avg = nabil['Close Price'].mean()
ntc_avg = ntc['Close Price'].mean()
sanima_avg = sanima['Close Price'].mean()
smata_avg = smata['Close Price'].mean()

bfc_traded_avg = bfc['Total Traded Shares'].mean()
nabil_traded_avg = nabil['Total Traded Shares'].mean()
ntc_traded_avg = ntc['Total Traded Shares'].mean()
sanima_traded_avg = sanima['Total Traded Shares'].mean()
smata_traded_avg = smata['Total Traded Shares'].mean()

# Create bar chart for average Close Price
companies = ['BFC', 'NABIL', 'NTC', 'Sanima', 'SMATA']
avg_close_prices = [bfc_avg, nabil_avg, ntc_avg, sanima_avg, smata_avg]

plt.figure(figsize=(10, 5))

# Bar chart for average Close Price
plt.bar(companies, avg_close_prices, color=['aqua', 'brown', 'grey', 'purple', 'pink'])
plt.title('Comparison of Average Close Prices (2000-2021)')
plt.xlabel('Company')
plt.ylabel('Average Close Price')
plt.show()

# Create bar chart for average Total Traded Shares
avg_traded_shares = [bfc_traded_avg, nabil_traded_avg, ntc_traded_avg, sanima_traded_avg, smata_traded_avg]

plt.figure(figsize=(10, 5))

# Bar chart for average Total Traded Shares
plt.bar(companies, avg_traded_shares, color=['aqua', 'brown', 'grey', 'purple', 'pink'])
plt.title('Comparison of Average Total Traded Shares (2000-2021)')
plt.xlabel('Company')
plt.ylabel('Average Total Traded Shares')
plt.show()
```
